```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer, util

# Load FAQ dataset
faq_data = pd.read_csv('/content/faq_bengali.csv')
questions = faq_data['question'].tolist()
answers = faq_data['answer'].tolist()

# Load pre-trained multilingual BERT model
model = SentenceTransformer('bert-base-multilingual-cased')

# Encode FAQ questions
question_embeddings = model.encode(questions, convert_to_tensor=True)

def get_response(user_query):
    # Encode user query
    query_embedding = model.encode(user_query, convert_to_tensor=True)

    # Compute cosine similarity
    cos_scores = util.cos_sim(query_embedding, question_embeddings)[0]

    # Find best match
    top_idx = torch.argmax(cos_scores).item()
    return answers[top_idx]

# Test the chatbot
while True:
    query = input("আপনার প্রশ্ন লিখুন (বন্ধ করতে 'exit' লিখুন): ")
    if query.lower() == 'exit':
        break
    response = get_response(query)
    print(f"উত্তর: {response}")
```
